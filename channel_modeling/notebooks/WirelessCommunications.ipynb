<h2> Resumen </h2>

Esta Jupyter Notebook tiene por objetivo servir como introducción a la comunicaciones

<h2>Propiedades de un sistema de comunicación inalámbrica</h2>

<ol>
    <li>El canal es un medio compartido, por lo que está muy presente el problema de interferencia.</li>
    <li>El canal es un medio que varía con el tiempo, por movimientos del usuario o cambios del ambiente.</li>
    <li>Tanto la potencia como la distribución en tiempo y frecuencia de la señal transmitida deben ser controladas 
        cuidadosamente para evitar molestar a otros sistemas.</li>
    <li>La señal recibida es de la forma:
        
        $$
            y_k = \sum_{l=-L_1}^{L_2}h_{l,k}x_{k-l} + w_k + i_k
        $$
        <ul>
            <li>$x_k$ son los símbolos transmitidos de la constelación utilizada.</li>
            <li>$w_k$ es el ruido aditivo gaussiano.</li>
            <li>$i_k$ es la interferencia de otros usuarios.</li>
            <li>$h_{l,k}$ es el canal, $l$ representa el delay (siendo $L_1+L_2$ el número total de coeficientes del canal) y 
                $k$ es la variación en el tiempo del canal.</li>
        </ul>
        Si se considera por ejemplo que $L_1=L_2=1$, entonces la señal recibida será de la forma:
        
        $$
            y_k = h_{-1,k}x_{k+1} + h_{0,k}x_{k} + h_{1,k}x_{k-1} + w_k + i_k
        $$
    </li>
</ol>

<strong>El canal de un sistema de comunicación inalámbrica genera interferencia intersímbolo, variabilidad en el tiempo e 
interferencia de otros usuarios.</strong>

<h3>Canal inalámbrico</h3>

La respuesta al impulso del canal depende de:

<ul>
    <li>Patrón de antena y ambiente: afectan a la amplitud.</li>
    <li>Ambiente: afecta al retardo. La diferencia de tiempo entre la primer y última componente de señal que arriba al receptor se denomina <strong>retardo de propagación (delay spread)</strong>.</li>
    <li>Movilidad del transmisor, receptor y cambios en el ambiente: generan la variación en el tiempo. El tiempo durante el cual el canal 
        permanece aproximadamente igual se denomina <strong>tiempo de coherencia</strong>.</li>
</ul>

El canal puede ser aproximado mediante 3 componentes: 
<ol>
    <li>Pérdidas de camino (path loss): debido al decaimiento de la potencia con la distancia.</li>
    <li>Sombreado (shadowing): debido a objetos de gran tamaño.</li>
    <li>Multicamino (multipath): debido a pequeños cambios en el ambiente.</li>
</ol>

Matemáticamente la respuesta real pasabanda puede expresarse como una superposición de $K$ caminos, cada uno de los cuales posee una amplitud y un delay que también pueden variar con el tiempo:

$$
        h_{PB}(t,\tau)=\sum_{k=1}^K A_k(t)\delta(\tau-\tau_k(t))
$$

La respuesta compleja equivalente en banda base es:

<center>  
<img src="images/PbBbConversion.png"  height="600" width="600"/>
</center> 
    
$$
        h_{BB}(t,\tau)=\sum_{k=1}^K A_k(t)exp(-j2\pi f_c\tau_k(t))\delta(\tau-\tau_k(t))
$$
<ul>
    <li>Los delays varían por cambios en la longitud de los caminos de propagación de la señal.</li>
    <li>La frecuencia varía por el efecto Doppler.</li>
    <li>La amplitud varía por cambios en el ambiente.</li>
</ul>

<h2>Propagación de ondas de radio</h2>

La antena dipolo de media onda es una de las antenas más ampliamente utilizadas en frecuencias arriba de $2\,\text{MHz}$. Es una antena formada por dos conductores de longitud total igual a la mitad de una longitud de onda. Una de las características de la misma es que la amplitud de la corriente a lo largo del dipolo tiene forma sinusoidal

<center> 
<img src="images/dipoleAntenna2.gif"  height="400" width="400"/>
</center> 

Otra característica de este tipo de antenas es que es posible observar como cerca de la antena (en el campo cercano) el patrón de radiación es curvo. Mientras que a una distancia mucho mayor de la antena (en el campo lejano), el patrón de radiación es plano.

<center> 
<img src="images/dipoleAntenna.gif"  height="400" width="400"/>
</center> 

El patrón de radiación en el campo lejano puede ser direccional o no direccional, dependiendo del tipo de antena, y permite utilizar la teoría de ondas planas. Lo cual posibilita la aplicación de un modelo más simplificado de radiación: el <strong>modelo basado en rayos</strong>. El mismo establece que los rayos son lanzados desde el transmisor al receptor y son perpendiculares a las ondas planas.

<h3>Modelo de radiación basado en rayos</h3>

Bajo este modelo de radiación, si consideramos que un rayo se encuentra con un obstáculo, pueden generarse distintos efectos según la interacción de este rayo con el obstáculo.

<ul>
    <li>Transmisión: cuando el objeto permite la transmisión del rayo a través de el.</li>
    <li>Reflexión: cuando las irregularidades de la superficie son pequeñas en comparación con la longitud de onda de la onda incidente, el objeto es relativamente plano.</li>
    <li>Dispersión: cuando las irregularidades de la superficie son de un orden de magnitud comparable al tamaño de la longitud de onda de la onda incidente.</li>
    <li>Difracción: cuando el rayo cambia de dirección alrededor de las esquinas de un obstáculo.</li>
</ul>

<strong>UN CAMINO DE PROPAGACIÓN</strong> (Free space pathloss)

Si se considera un modelo de transmisión simple en el que sólo existe un camino de propagación de onda (la línea de vista LOS), la señal recibida en el receptor, sin considerar el ruido, es de la forma:

$$
    r(t)= \Re\left\{\frac{\lambda\sqrt{G_0}e^{-j2\pi d/\lambda}}{4\pi d}u(t)e^{j2\pi f_ct}\right\}
$$

<ul>
    <li>$u(t)$ es la señal transmitida.</li>
    <li>$e^{j2\pi f_ct}$ representa la modulación de la señal con la frecuencia de portadora $f_c$.</li>
    <li>$e^{-j2\pi d/\lambda}$ evidencia el efecto de rotación del canal generado por la propagación con la distancia.</li>
    <li>$1/d$ evidencia el efecto de decaimiento del canal, el cual genera que la potencia disminuya con la distancia.</li>
    <li>$\lambda\sqrt{G_0}$ evidencia el efecto de la antena:
        <ul><li>$G_0$ producto de las ganancias de la antena transmisora ($G_t$) y receptora ($G_r$) en la dirección de línea de vista.</li>
        <li>$\lambda$ es la longitud de onda de la portadora y es función de la apertura de la antena (valores de area equivalentes 
            que permiten describir las características de captura de potencia de una antena).</li>
        </ul>   
    </li>
</ul>

<center> 
<img src="images/los.jpg"  height="500" width="500"/> 
</center> 

La potencia de la señal recibida a una distancia $d$ puede calcularse a partir de la <strong>FÓRMULA DE FRIIS</strong> como: 

$$
    P_{RX}(d)= P_{TX}G_{TX}G_{RX}\left(\frac{\lambda}{4\pi d}\right)^2
$$

comunenmente tambien escrita como

$$
\frac{P_{RX}(d)}{P_{TX}}= G_{TX}G_{RX}\left(\frac{\lambda}{4\pi d}\right)^2
$$

Retomando la primer froma y aplicando logaritmos para considerar las 
cantidades en dB:

$$
    P_{RX}(d)\text{[dBm]}= P_{TX}\text{[dBm]} + G_{TX}\text{[dB]} + G_{RX}\text{[dB]}+ 20\log_{10}(\lambda)-20\log_{10}(4\pi)-20\log_{10}(d)
$$

<ul>
    <li>$P_{TX}$ potencia transmitida de la señal $u(t)$.</li>
    <li>$G_{TX}$ ganancia de la antena transmisora en la dirección de LOS.</li>
    <li>$G_{RX}$ ganancia de la antena receptora en la dirección de LOS.</li>
    <li>$\left(\lambda/(4\pi d)\right)^2$ <strong>factor de pérdida en espacio libre</strong></li>
</ul>

**La potencia recibida decrece con el cuadrado del incremento de la frecuencia de portadora.**

**La potencia de la señal recibida decrece inversamente con el cuadrado de la distancia entre las antenas transmisora y receptora.**

__Ejercicio 1 Capitulo 2__ 

Considerando un modelo de pérdidas por espacio libre, encontrar la potencia trasmitida requerida
par obtener una potencia recibida de $1\text{dBm}$ para un sistema inalámbrico con antenas isotrópicas 
($G_t=G_r=1$) y una frecuencia de portadora $f_c=5 \text{GHz}$, asumiendo una distancia $d=10\text{m}$. Repetir
para $d=100\text{m}$.

```python
# velocidad de luz [m/s]
c = 300_000_000

# frecuencia de la portadora [Hz]
fc_ej1 = 5_000_000_000

# longitud de onda [m]
l_ej1 = c / fc_ej1

# potencia recibida [W]
pr_ej1 = 0.001

# calculando Pt
def print_answer_ej1(d = 10):
    pt = pr_ej1 / (l_ej1 / (4 * np.pi * d))**2
    print(f'Para una distancia de {d} m la potencia trasmsititda debe ser: {(pt / 1000):.2f}Kw')

widgets.interact(print_answer_ej1, d = (10, 100, 10));
```

<strong>DOS CAMINOS DE PROPAGACIÓN</strong> (Two-Ray Multipath model)

Si se considera un modelo de transmisión en el que existen dos caminos de propagación de onda 
(la línea de vista LOS y alguna reflexión).
Este modelo comumente es usado cuando un simple terreo u otra reflección
domina el efecto multicamino. La señal recibida consiste de dos componentes: el LOS que corresponda a la 
señal transmitida que se propaga a traves del espacio libre y una componente de reflección que es la que 
comunmente se refleja del terreno.
La señal recibida en el receptor, sin considerar el ruido, es de la forma:

$$
    r_{2ray}(t)= \Re\left\{\frac{\lambda}{4\pi}\left[\frac{\sqrt{G_0}u(t)e^{j2\pi d_0/\lambda}}{d_0}+\frac{R\sqrt{G_1}u(t-\tau)e^{j2\pi (d_{11}+d_{12})/\lambda}}{d_{11}+d_{12}})\right]e^{j2\pi f_c t}\right\}
$$

<ul>
    <li>$u(t-\tau)$ el retardo causado por la diferencia de longitud entre los caminos de propagación afecta a la señal recibida.</li>
    <li>$e^{j2\pi d_0/\lambda}$ y $e^{j2\pi (d_{11}+d_{12})/\lambda}$ cada camino tiene su propio efecto de rotación que depende de la longitud del mismo.</li>
    <li>$d_0$ longitud del camino de linea de vista</li>
    <li>$d_1=d_{11}+d_{12}$ longitud del camino reflejado.</li>
    <li>$G_0$ producto de las ganancias de las antenas en la dirección de línea de vista.</li>
    <li>$G_1$ producto de las ganancias de las antenas en la dirección de reflexión.</li>
    <li>R coeficiente de reflexión de la superficie.</li>
</ul>

<center> 
<img src="images/two-ray.png"  height="700" width="600"/> 
</center> 

En la siguiente figura puede observarse la variación de la potencia recibida con la distancia. La potencia recibida en general disminuye con la distancia, pero posee una zona de variación con un patrón irregular. Esto se debe a la interferencia aditiva o destructiva entre los dos caminos.

<center> 
<img src="images/twoRay.png"  height="700" width="500"/> 
</center>


* La señal recibida consiste de dos componentes:
    * Una componente LOS, que se propaga por el espacio libre.
    * Una componente reflejada, que es la señal transmitida reflejada en el suelo.
* Las componentes de señal recibida recorren cada una una distancia distinta.
* El retardo de propagación (delay spread) del modelo es la diferencia entre el retardo de las componentes de señal recibida: $\tau_1-\tau_0 = \frac{\left(d_1-d_0\right)}{c}$

Si el ancho de banda de la señal transmitida es pequeño comparado con el retardo de propagación, es decir que $(\tau_1-\tau_0) \ll B_u^{-1}$, siendo $\tau_1$ y $\tau_0$ los retardos de cada una de las componentes de la señal, la potencia recibida $P_r$ se calcula mediante la siguiente fórmula:

$$
    P_r = P_t \left(\frac{\lambda}{4\pi}\right)^2\left|\frac{\sqrt{G_0}}{d_0}+\frac{R\sqrt{G_1}e^{-j\Delta\phi}}{d_1}\right|^2
$$

- $P_t$ es la potencia de la señal transmitida.
- $G_0$ es el producto de los patrones de radiación de las antenas en la dirección LOS.
- $G_1$ es el producto de los patrones de radiación de las antenas correspondientes a la reflección de suelo.
- $R$ es el coeficiente de reflección del suelo.
- $d_0$ y $d_1$ son las distancias recorridas por la componente LOS y reflejada de la señal.
- $\Delta\phi=2\pi(d_1-d_0)/\lambda$ es la diferencia de fase entre las dos componentes de señal recibidas.


**Para valores de $d$ muy grandes, $d_1=d_0$, $G_1=G_0$ y la potencia recibida es inversamente proporcional a $d^4$ e independiente de la longitud de onda $\lambda$.**

Al graficar la potencia recibida con este modelo, se obtienen tres secciones diferenciadas:
* Para distancias pequeñas $(d < h_t)$ las componentes se suman contructivamente a medida que $d$ aumenta.
* Cuando $\Delta\phi=\pi$ se alcanza un máximo, luego del cual las componentes de señal experimentan interferencias constructivas y destructivas, observandose máximos y mínimos sucesivos.
* En la denominada distancia crítica $d_c=4h_th_r/\lambda$ se alcanza el último máximo. Luego del mismo la potencia de la señal decrece proporcionalmente a $d^{-4}$.



__Ejercicio 2 Capitulo 2__

Considerando un modelo de 2 rayos con una separación entre el transmisor y receptor $d=100\text{m}$, $h_t=10\text{m}$ y $h_r=2\text{m}$, encontrar el retardo de propagación entre ambas señales (delay spread).

```python
def print_answer_ej2(d = 100, ht = 10, hr = 2):
    num = np.sqrt((ht + hr)**2 + d**2) - np.sqrt((ht - hr)**2 + d**2)
    delay_spread = num / c
    print(f'El retardo de propagación para d = {d} m, ht = {ht} m y hr = {hr} m ' 
          f'es {(delay_spread * 10**9):.2f}ns')
    
widgets.interact(print_answer_ej2, d = (10, 1000, 10), ht = (5, 100, 5), hr = (1, 10, 1));   
```

<strong>AMBIENTES DE PROPAGACIÓN COMPLEJOS</strong>

El trazado de rayos ya no es posible de implementar, por lo que se utiliza otro tipo de modelo que considere 3 efectos presentes en el canal: 

<ol>
    <li><strong>Pérdidas de camino (path loss)</strong>: genera disminución de la potencia recibida con la distancia.</li>
    <li><strong>Sombreado (shadowing)</strong>:genera variaciones de la potencia recibida con grandes variaciones de distancia. Debido a bloqueos de la señal.</li>
    <li><strong>Multicamino (multipath)</strong>: genera pequeñas variaciones de la potencia recibida con pequeñas variaciones de distancia (pocas longitudes de onda). Debido a reflexión, dispersión y difracción.</li>
</ol>

<center>

</center>

$$
P_r\text{[dBm]} = P_t\text{[dBm]}+ \text{pathloss[dBm]}+\text{shadowing[dBm]} + \text{multipath[dBm]}
$$

Para caracterizar correctamente el comportamiento del canal se utiliza un tipo de modelo particular para analizar cada efecto en particular:
<ol>
    <li>Pérdidas de camino (path loss): modelo determinístico.</li>
    <li>Sombreado (shadowing): modelo aleatorio de tipo log-normal.</li>
    <li>Multicamino (multipath): modelo aleatorio.</li>
</ol>

```python
# parametros

# frecuencia
f_ej8 = 900e6

# velocidad de la luz
c = 3e8

# producto de los patrones de radiacion de las antenas en la direccion LOS
Gl = 1

# coeficiente de reflexion del suelo
R = -1
count = 1

d_ej8 = np.arange(1, 1000001)

# definiendo funcion para calcular la curva de interes

def make_plot(Gr, ht_ej8, hr_ej8, f_ej8):
    # d0: distancia LOS 
    l = np.sqrt(d_ej8**2 + (ht_ej8 - hr_ej8)**2)
    # d1: disntancia del camino reflejado
    r = np.sqrt(d_ej8**2 + (ht_ej8 + hr_ej8)**2)
    # long de onda
    lambd_ej8 = c / f_ej8 
    # diferencia de fase entre dos componentes de señal recibida 
    phd = 2 * np.pi / lambd_ej8 * (r - l)
    # distancia critica: se alcanza el ultimo maximo local y Pr decae con d^-4
    dc = 4 * ht_ej8 * hr_ej8 / lambd_ej8
    vec = (np.sqrt(Gl) / l) + (R * np.sqrt(Gr) / r) * np.exp(1j * phd)
    return (lambd_ej8 / (4 * np.pi))**2 * (np.abs(vec))**2


# configurando widgets

slider_Gr = widgets.SelectionSlider(
    options = [('0 dB', 1), ('-5 dB', 0.316), ('-10 dB', 0.1), ('-20 dB', 0.01)],
    description = 'Gr (dB)',
    value = 1
)

slider_f = widgets.FloatSlider(
    min = 0.1,
    max = 5,
    step = 0.1,
    value = 0.9,
    description = 'f (GHz)'
)


slider_hr = widgets.FloatSlider(
    min = 5,
    max = 100,
    step = 5,
    value = 15,
    description = 'hr (m)'
)


slider_ht = widgets.FloatSlider(
    min = 5,
    max = 100,
    step = 5,
    value = 50,
    description = 'ht (m)'
)


out_ej8 = widgets.Output()

with out_ej8:
    fig, ax = plt.subplots(figsize = (6, 4))
    fig.suptitle('Potencia vs Distancia')
    pr = make_plot(slider_Gr.value, slider_ht.value, slider_hr.value, slider_f.value * 1e9)
    fig.canvas.header_visible = False
    fig.canvas.footer_visible = False
    line, = ax.plot(10 * np.log10(d_ej8), 10 * np.log10(pr) - 10 * np.log10(pr[0]),  label = 'Pr')
    lambd_ej8 = c / (slider_f.value * 1e9)
    dnew = np.arange(4 * slider_ht.value * slider_hr.value / lambd_ej8, 1000001)
    ax.plot(10 * np.log10(dnew),-20 * np.log10(dnew) + 50, label = r'$d^{-2}$')
    ax.plot(10 * np.log10(dnew),-40 * np.log10(dnew) + 100, label = r'$d^{-4}$')
    ax.set_xlabel(r'$10 \log_{10}(d)$')
    ax.set_ylabel(r'$10 \log_{10}(Pr)$')
    ax.set_xlim(0, 70)
    ax.set_ylim(-120, 20)    
    ax.grid(True)
    ax.legend()

out2_ej8 = widgets.Output()


out2_ej8 = widgets.HTMLMath(value = f'La distancia critica en escala logaritmica ' 
                            f'{(10 * np.log10(4 * slider_ht.value * slider_hr.value / (c / (slider_f.value * 1e9)))):.2f}')
    
out3_ej8 = widgets.HTMLMath(value = r'Cuando $G_r$ disminuye, el comportamiento '
                            r'asintótico <br> de $P_r$ tiende a $d^{-2}$. Esto tiene '
                            r'sentido ya que el efecto <br> del camino reflejado es reducido '
                            r'y se parece más <br> a tener un solo camino LOS.')

grid_ej8 = widgets.GridspecLayout(1, 2)

grid_ej8[0, 0] = widgets.VBox([slider_f, slider_Gr, slider_ht, slider_hr, out2_ej8, out3_ej8])
grid_ej8[0, 1] = out_ej8

# obtencion de las curvas usando la formula 2.13 del libro para el modelo de multicamino de dos rayos


# funcion para actualziar el grafico

def update_plot(change):
    pr = make_plot(slider_Gr.value, slider_ht.value, slider_hr.value, slider_f.value * 1e9)
    line.set_ydata(10 * np.log10(pr) - 10 * np.log10(pr[0]))
    fig.canvas.draw()
    out2_ej8.value =  f'La distancia critica en escala logaritmica {(10 * np.log10(4 * slider_ht.value * slider_hr.value / (c / (slider_f.value * 1e9)))):.2f}'
    
    

# vinculando widgets

slider_Gr.observe(update_plot, 'value')
slider_ht.observe(update_plot, 'value')
slider_hr.observe(update_plot, 'value')
slider_f.observe(update_plot, 'value')

# visualizando

grid_ej8 
```

<h2>Probabilidad de corte</h2>
La probabilidad de corte (outage probability) es la probabilidad de que la potencia recibida sea menor a algún umbral. Este umbral generalmente está relacionado con la potencia mínima requerida por el receptor para capturar la información proveniente del transmisor.

$$
    P_{out}= p(P_r < P_{min})
$$

<ol>
    <li><strong>Unicamente existen pérdidas de camino</strong>: La potencia disminuye con la distancia desde el transmisor, por lo que a partir de algún punto la potencia recibida comenzará a ser menor al umbral y la probabilidad de corte será igual a 1. Para todas las distancias menores a este punto, la potencia recibida será mayor al umbral, por lo que la probabilidad de corte será 0.</li>
    <li><strong>Existen pérdidas de camino y sombreado</strong>:
        
    $$
    \begin{align}
        P_{out}(d) &= p(P_r(d) < P_{min})\\
                   &= p(P_r(d)\text{[dBm]} < P_{min}\text{[dBm]})\\
                   &=p(P_t\text{[dBm]} + K\text{[dB]}-10\gamma \log_{10}\left(\frac{d}{d_0}\right)+\psi_{dB} < P_{min}\text{[dBm]})
    \end{align}  
    $$
   
    <br>       
    $\psi_{dB}$ es la única variable aleatoria de tipo gaussiana $\psi_{dB} \sim \mathcal{N}(0,\sigma_{\psi_{dB}}^2)$, por lo que esta probabilidad se puede resolver haciendo uso de la función $\mathcal{Q}$:  
    <br>  
    $$
    P_{out}(d) = 1 - Q \left(\frac{P_{min}\text{[dBm]}-(P_t\text{[dBm]} + K\text{[dB]}-10\gamma \log_{10}(d/d_0)}{\sigma_{\psi_{dB}}}\right)
    $$
    <br>         
    La probabilidad de corte será muy pequeña (cercana a 0) cuando el receptor se encuentre muy cerca del transmisor y mientras se vaya alejando del mismo la probabilidad irá incrementando hasta llegar a 1.
    </li>
</ol>
